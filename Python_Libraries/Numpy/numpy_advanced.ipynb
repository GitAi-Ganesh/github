Numpy arrays vs Python List

```python
# speed
# list
a = [i for i in range(10000000)]
b = [i for i in range(10000000,20000000)]

c = []
import time 

start = time.time()
for i in range(len(a)):
  c.append(a[i] + b[i])
print(time.time()-start)
```

```python
#numpy 
import numpy as np 
np.arange (10000000)
b = np.arange(10000000,20000000)

start = time.time()
c = a + b
print(time.time()-start) #it take very less time to execute
```

```python
# memory
a = [i for i in range (10000000)]
import sys
#library for system to know the size of this list in the memory

sys.getsizeof(a)
```

```python
a = np.arange(10000000,dtype=np.int8)#as we can reduce size in np by using dtype
sys.getsizeof(a)
```

Advance indexing

```python
# Normal Indexing and slicing

a = np.arange(24).reshape(6,4)
a
```

```python
a[1,2]
```

```python
a[1:3,1:3]
```

```python
# Fancy Indexing

a[:,[0,2,3]]
```

```python
# Boolean Indexing
a = np.random.randint(1,100,24).reshape(6,4)
a
```

```python
# Boolean Indexing
import time
b= time.time()
a = np.random.randint(1,100,24).reshape(6,4) #random integers

print (time.time()-b)

a
```

```python
# find all numbers greater than 50
a[a > 50]
```

```python
# find out even numbers
a[a % 2 == 0]
```

```python
# find all numbers greater than 50 and are even
#here we merge two  different logics

a[(a > 50) & (a % 2 == 0)]
```

```python
# find all numbers not divisible by 7
a[~(a % 7 == 0)]
```

### Broadcasting

The term broadcasting describes how NumPy treats arrays with different shapes during arithmetic operations.

The smaller array is “broadcast” across the larger array so that they have compatible shapes.
