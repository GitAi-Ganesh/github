Numpy arrays vs Python List

```python
# speed
# list
a = [i for i in range(10000000)]
b = [i for i in range(10000000,20000000)]

c = []
import time 

start = time.time()
for i in range(len(a)):
  c.append(a[i] + b[i])
print(time.time()-start)
```

```python
#numpy 
import numpy as np 
np.arange (10000000)
b = np.arange(10000000,20000000)

start = time.time()
c = a + b
print(time.time()-start) #it take very less time to execute
```

```python
# memory
a = [i for i in range (10000000)]
import sys
#library for system to know the size of this list in the memory

sys.getsizeof(a)
```

```python
a = np.arange(10000000,dtype=np.int8)#as we can reduce size in np by using dtype
sys.getsizeof(a)
```
