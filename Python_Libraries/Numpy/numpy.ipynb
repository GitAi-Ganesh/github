Numpy Array

```python
# np.array
import numpy as np
a = np.array([1,2,3])
print(a)
#output is a vector
```

```python
#this is a 2D array
b = np.array([[1,2],[2,3]])
print(b)
#output is a matrix
```

```python
#3D array
c= np.array([[[1,2],[2,3]],[[3,4],[4,5]]])
print(c)
#output is a Tensor
```

```python
#dtype
e = np.array([1,2,3],dtype=float)
print(e)
#here the datatype us changes as float it can be bool, int, string
```

```python
#np.arange

a =np.arange(1,10) #it gives numbers from 1 to 10

b =np.arange(1,10,2) #it gives numbers leaving every 2 space
print(a,b)
```

```python
# np with reshape

np.arange(1,11).reshape(5,2)
#it will form a 5 rows and 2 column
```

```python
#np.ones and np.zeros
#this are use for initializing arrays  
#specially used in neural network and machine learning
np.ones((3,4))
```

```python
np.zeros((3,4))
```

```python
#np.random

np.random.random((3,4))
```

```python
#np.linspace

np.linspace(1,10,5)

#first number(1)is starting point
#second number(10) is ending point
#third number (5) is point required on equal space
```

```python
#np.identity

np.identity(8)# here in identity only diagonals are 1 remaining are 0
```

Array Attributes

```python
a1 = np.arange(10) #1 dimension
a2 = np.arange(12,dtype=float).reshape(3,4) #2 dimension
a3 = np.arange(8).reshape(2,2,2) # 3 dimension
```

```python
#ndim (number of Dimensions)

a1.ndim
```

```python
#shape 
print(a3.shape)
#use for knowing number of column and shapes
```

```python
#size
print(a2.size)
#number of items
```

```python
# itemsize
a1.itemsize
```

```python
#dtype
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```

Changing Datatype

```python
#astype
#it is use to change the datatype to reduce the memory size of the datatype

a1.astype(np.int32)

```

Array function

```python
a1 = np.random.random((3,4))
a1 = np.round(a1*100)
a1
```

```python
#max, min, sum, prod

np.max(a1)
```

```python
np.min(a1)
```

```python
np.sum(a1)
```

```python
np.prod(a1,axis=0) #axis means which column 0 column and 1 is row
```

```python
     # dot product
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(12,24).reshape(4,3)

np.dot(a2,a3)
```

```python
# log and exponents
np.exp(a1)
```
